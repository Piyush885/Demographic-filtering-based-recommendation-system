```python
import pandas as pd
```

```python
df1=pd.read_csv('movie_metadata.csv')
```

```python
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1=df1.dropna()
```

```python
df1.isnull().sum()
```

```python
#First we will find best recommended movie based on demographic filtering---This method is simple. This will recommend the most imdb rated movies
```

```python
cutoff=df1['imdb_score'].quantile(0.9)
```

```python
#So our cutoff is 77percentile. We will remove all the movies less than this percentile
```

```python
df1=df1[df1.imdb_score>cutoff]
```

```python
df1.head()
```

```python
#now we will print top 10 most rated movies
```

```python
df1=df1.sort_values('imdb_score',ascending=False)
```
